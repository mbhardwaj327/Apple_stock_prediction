```python
df['date']=pa.to_datetime(df.date,format="%Y/%m/%d")
```

```python
df.head()
```

```python
df.info()
```

```python
X=df.loc[:,['date','close']]
```

```python
X.head()
```

```python
X.columns=['ds','y']
X.tail()
```

```python
from fbprophet import Prophet
```

```python
pr=Prophet()
```

```python
pr.fit(X)
```

```python
pr.component_modes
```

```python
Futuredata=pr.make_future_dataframe(365,freq='D')
```

```python
Futuredata.tail()
```

```python
predict=pr.predict(Futuredata)
```

```python
predict.tail()
```

```python
pr.plot(predict)
```

```python
pr.plot_components(predict)
```

```python
from fbprophet.diagnostics import cross_validation
```

```python
df_cv=cross_validation(pr,horizon="365 days",period="180 days",initial="730 days")
```

```python
df_cv.head()
```

```python
df_cv.tail()
```

```python
from fbprophet.diagnostics import performance_metrics
```

```python
df_perform=performance_metrics(df_cv)
```

```python
df_perform.head()
```

```python
from fbprophet.plot import plot_cross_validation_metric
```

```python
flg=plot_cross_validation_metric(df_cv,metric='rmse')
```
